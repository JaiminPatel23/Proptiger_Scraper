```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import json
import io
```

```python
# Make a request to https://www.proptiger.com/mumbai/property-sale
URL = 'https://www.proptiger.com/mumbai/property-sale'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')

prettyHTML = soup.prettify()
print(prettyHTML)
```

```python
# Extract body of page
page_body = soup.body
print(page_body)
```

**Here, we will extract the following data :-**

1) Page Title

2) Data Last Updated

3) City Name

4) Count The Number Of Projects In Last Six Months

5) Total Projects 

6) Image Detail

7) Project Name

8) Project Address

9) Project Price

10) Project Builder

11) Possession

12) RERA ID

13) View Details

14) Get Callback

```python
# Extract title of page - <title>(...)</title>
def extract_title(soup):
    title = soup.find('title')
    print('TITLE -')
    print()
    print(title.text)


# Extracting data to get basic overview of the page
def overview(soup):

    last_update = soup.find('div', class_ = 'lastUpdated').text

    cityname = soup.find('h2', class_ = 'heading').text
    
    perc_time = soup.find('div', class_ = 'trend green').text
    projects = perc_time.replace('%','% ')
    
    total_projects = soup.find('li', class_ = 'total-projects').text.strip()
    
    print()
    print()
    print()    
    print()
    print('OVERVIEW -')
    print()
    print(last_update)
    print(cityname)
    print(projects)
    print(total_projects)


extract_title(soup)
overview(soup)
```

```python
# Extracting data to get filters available on the page
def filters(soup):

    filter = soup.find('div', class_ = 'clearfix max1028')
    filters = filter.find('div').text.strip()
    sorted_by = filter.find('section').text.strip()

    print('FILTERS -')
    print()
    print(filters)
    print()    
    print()
    print()
    print('SORTED BY -')
    print()
    print(sorted_by)


filters(soup)
```

```python
# TILL THE FIRST MOUSE SCROLL - FIRST 15 PROPERTIES (PAGE MOVES DOWN TO NEXT SET OF 15 PROPERTIES AND SO ON TILL TOTAL PROJECTS COUNT IS COVERED)

container = soup.find_all("section", {"class":"project-card-main-wrapper"})  
print("Number of Properties - ", len(container))
print()
print()

for i in container:

    det1 = i.find_all("div", class_ = "proj-name")         
    try:
        project_name = det1[0].find('span')
    except:
        project_name = "n/a"
    print("Project Name - ", project_name.text.strip())
    

    det2 = i.find_all("span", {"itemprop":"address"})#or by ("div", {"class":"loc"})
    try:
        project_loc  = det2[0]["title"]# or by .text
    except:
        project_loc = "n/a"
    print("Project Location - ", project_loc)


    det3 = i.find_all("div", {"class":"project-price"})
    try:
        price_range = det3[0].text.strip()
    except:
        price_range = "n/a"
    print("Project Price Range - ", price_range)


    det4 = i.find_all("div", {"class":"projectBuilder put-ellipsis"})
    try:
        project_builder = det4[0].text.strip()
    except:
        project_builder = "n/a"
    print("Project Builder - ", project_builder)
    

    det5 = i.find_all("div", class_ = 'possession-wrap')
    try:
        project_poss = det5[0].text.strip()
    except:
        project_poss = "n/a"
    print("Project Possession - ", project_poss)
    
  
    det7 = i.find_all("div", {"class":"rera-id put-ellipsis"})     
    try:
        registered_id = det7[0].text.strip()
    except:
        registered_id = "n/a"
    print("Registered Identification - ", registered_id)


    base_url = "https://www.proptiger.com"
    hyper_url = i.div['data-url']
    try:
        url = base_url+hyper_url
    except:
        url = "n/a"
    print("Link for More Details - ", url)


    print()
```
